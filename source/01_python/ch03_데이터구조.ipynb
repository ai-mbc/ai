```python
def safe_index(lst,item):
    '''
    첫번째매개변수lst에서 item요소가있는index를반환하는함수
    item요소가없으면-1을반환
    '''
    if item in lst:
        return lst.index(item)
    else:
        return -1
```

```python
def safe_remove(lst, remove_item) :
    '''
    첫번째 매개변수 lst에서 remove_item 데이터 제거
    해당 remove_item이 lst에 없으면 메세지 출력
    '''
    if remove_item in lst:
        lst.remove(remove_item)
    else:
        print('해당 데이타는 없는 데이타입니다')
```

```python
names=['길동','길자','순자']
tels=['010-3344-8483','010-2334-3322', '010-3848-2882']
adds=['서울','인천','판교']
list(enumerate(zip(names,tels,adds)))
for idx, (name,tel,add) in enumerate(zip(names,tels,adds)):
    print(f'{idx+1}. {name}의 연락처는 {tel}이고, {add}에 살아요')
```

```python
#names,tels,adds를 딕셔너리 리스트로 만들기
members=[]
for name,tel,add in zip(names,tels,adds):
    member={'name':name, 'tel':tel, 'address':add}
    members.append(member)
members
```
