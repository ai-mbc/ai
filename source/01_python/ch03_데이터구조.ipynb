## 1.4 인덱싱에서 요소 인덱스 찾기
- 변수[idx]:인덱스를이용한접근(없는인덱스의경우에러남)
- count(찾을데이타):해당찾을데이타가나오는갯수반환
- index(찾을데이터):0번째부터해당찾을데이타가 몇번째인덱스에 있는지반환
- index(찾을데이터,start):start번째부터 해당 찾을데이터가 몇번째 인덱스에 있는지반환

```python
numbers=[1,2,3,[1,2,3],1,2,3]
```

```python
numbers.count(2)
```

```python
numbers.count(9)
```

```python
if numbers.count(1):
    print(numbers.index(1))
else:
    print('찾을수없음')
```

```python
if 9 in numbers:
    print(numbers.index(9))
else:
    print('찾을수없음')
```

```python
def safe_index(lst,item):
    '''
    첫번째매개변수lst에서 item요소가있는index를반환하는함수
    item요소가없으면-1을반환
    '''
    if item in lst:
        return lst.index(item)
    else:
        return -1
```

```python
def safe_remove(lst, remove_item) :
    '''
    첫번째 매개변수 lst에서 remove_item 데이터 제거
    해당 remove_item이 lst에 없으면 메세지 출력
    '''
    if remove_item in lst:
        lst.remove(remove_item)
    else:
        print('해당 데이타는 없는 데이타입니다')
```

```python
#데이타 구조의 복제(x)
numbers=[5,1,9]      #원본
new_numbers=numbers #복제본
print(f'원본i={numbers},복사본i={new_numbers} / 원본주소:{id(numbers)},복사본주소:{id(new_numbers)}')
new_numbers[0]=555
print(f'원본i={numbers},복사본i={new_numbers} / 원본주소:{id(numbers)},복사본주소:{id(new_numbers)}')
```

```python
#복제(=깊은복사)
numbers=[5,1,9]
new_numbers=numbers.copy()
print(f'원본i={numbers},복사본i={new_numbers} / 원본주소:{id(numbers)},복사본주소:{id(new_numbers)}')
new_numbers.sort()
print(f'원본i={numbers},복사본i={new_numbers} / 원본주소:{id(numbers)},복사본주소:{id(new_numbers)}')
```
