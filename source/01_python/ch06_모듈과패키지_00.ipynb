```
파이썬 설치하는 방법
    - python.org/microsoft store : 파이썬 표준모듈200여개가 포함된 파이썬 인터프리터를 install(sys, re, datetime, time...)

    - anaconda : 몇백개 이상의 패키지, 개발도구(IDE tools;jupyter notebook)를 포함한 인터프리터를 설치(200명이상의 기업체에서는 유료)
    - uv(pip-파이썬 패키지 설치도구, 패키지/프로젝트 관리도구)를 이용한 설치

            - powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex(powershell관리자권한)

        * Set-ExecutionPolicy RemoteSigned -Scope CurrentUser

```

````python
%%time
import time
start=time.time()
cnt=0
while cnt<=10000000:
    cnt+=1
end=time.time()
print('반복문 1천만번 돈 시간 :',(end-start),'초')
````
