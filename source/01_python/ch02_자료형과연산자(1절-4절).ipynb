## 4.2 format()문법
- "{[인덱스]:[공백대체문자][정렬방식][자릿수][타입]}메시지".format(변수명)
 * 인덱스:format함수안의매개변수 순서
 * 공백대체문자는 정렬방식과 자릿수를 명시하여야 합니다. 자릿수를 여유롭게해야 합니다
 * 정렬방식:<(왼쪽정렬),>(오른쪽정렬),^(가운데정렬)
 * 정수자릿수:출력할 자릿수
 * 실수자릿수:전체출력될 자릿수, 변수중 출력할자릿수
 * 타입:d(10진수),f(실수4.1f:1은 소숫점자리수),s(문자),o(8진수),x(16진수),b(2진수)

````python
name=input('이름은?')
age=int(input('나이는?'))
````

````python
print('{}살 {}님 안녕하세요'.format(age,name))
print(f'{age}살 {name}님 안녕하세요')
````

````python
print('[{0:10}]살 [{1:10}]님 안녕하세요'.format(age,name))
print(f'[{age:10}]살 [{name:10}]님 안녕하세요')
````

````python
print('[{0:*<10}]살 [{1:_>10}]님 안녕하세요'.format(age,name))
print(f'[{age:*<10}]살 [{name:_>10}]님 안녕하세요')
````

````python
print('[{0:*<10d}]살 [{1:_>10s}]님 안녕하세요'.format(age,name))
print(f'[{age:*<10.1f}]살 [{name:_>10.2s}]님 안녕하세요')
````

````python
print('[{0:*^+10d}]살 [{1:_>10s}]님 안녕하세요'.format(age,name))
print(f'[{age:*<10.1f}]살 [{name:_>10.2s}]님 안녕하세요')
````

````python
a=2.755
sign='+'
width=10
digits=3
print('[{:+10.3}]'.format(a))
print('[{:{}{}.{}}]'.format(a,sign,width,digits))
print('[{:{sign}{width}.{digits}}]'.format(a,sign=sign,width=width,digits=digits))
````
