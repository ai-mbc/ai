# 1절. 조건문(if)
```
if 조건1:
   조건1이 참일경우 수행할 명령어들(들여쓰기중요)
elif 조건2:
     조건1이 거짓이고, 조건2가 참일 경우 수행할 명령어들(elif는 필요한 만큼사용가능)
else:
   모든 조건이 만족하지 않을 때 수행할 명령어(else는 단독사용불가)
```

## 2.1 for-in문
```
for 임시변수 in 나열가능한자료:
    반복할문장들
else :
    나열가능한자료를다소요하여 for블록을 다 수행한후,최종한번실행할문장들
    for문안의 break를 수행할경우,else블록을수행하지않음(else는선택)
```

````python
for i in range(1,10,2): #[1,3,5,7,9]
    print(i,end='\t')
else:
    print('\nDone')
print('EN')
````

## 2.2 while문
```
while 조건 :
      조건이 참일경우반복할문장들
else:
      조건이 거짓일 경우 한번 실행할문자(else선택사항
```

````python
#1~100까지 짝수들의 합을 계산하고 출력하시오(while문)
tot=0
i=1
while i<=100:
    if i%2==0:
        tot+=i
    i+=1 #i변수 1증가
else:
    print('1~100까지 짝수의 합은',tot)
````

````python
#사용자에게 두 정수를 입력받아 작은수부터 큰수까지 홀수들의 합을 누적하여 출력하시오
first=int(input('첫번째정수는?'))
second=int(input('두번째정수는?'))
#둘중 어느 것이 작은수인지
#(1)두 수중 minNo, maxNo에 할당
if first<second:
    minNo=first
    maxNo=second
else:
    minNo=second
    maxNo=first

#minNo=fist if first<seond else second
#maxNo=first if first>seond else seond
#(2)minNo~maxNo까지 홀수누적하기
tot=0
for n in range(minNo,maxNo+1):
    if n%2: #if no%!=0
        tot+=n
print(f'{minNo}부터 {maxNo}까지 홀수의 합은 {tot}입니다')
````

````python
#10. 1~100 사이의 정수를 맞추는 프로그램을 구현하시오
'''
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계.입력된 숫자가 틀리면 UP/DOWN 메시지를 출력하고 숫자를 맞추면 축하메세지와 함꼐 종료합니다.(1~100사이의수를 알아맞추기) 맞출 때까지 반복
2단계.수를못맞추면 min~max를조정,숫자를 맞추면 축하메세지와 함꼐 종료
3단계.2단계의기회를5회로제한.5회의기회가지나면실패메세지뿌리고끝남
'''
import random
min_=1
max_=100
randomNum=random.randint(1,100)
print('컴퓨터가발생한임의의수 :',randomNum)
while True:
    inputNum=int(input('1~100 사이의수를 맞춰보세요 :'))
    if inputNum>100|inputNum<1:
         print('범위초과')
    elif inputNum>randomNum:
         print('Down↓')
    elif inputNum<randomNum:
         print('Up')
    else:
         print('추카추카')
         break
print('done')
````

````python
#중첩로프탈출 방법1:flag변수이용
flag=False
for a in range(0,3):
    for b in range(1,3):
        if a==b:
           print('a={}, b={} break'.format(a,b))
           flag=True
           break
        print(a,b)
    if flag:
        break
print('done')
````
