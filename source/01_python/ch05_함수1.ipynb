## 여러개값을 return

````python
def my_add_return(a,b):
    "매개변수로들어온a,b값의합을 return합니다 -docstring"
    total=a+b
    return total,a,b #여러개값을return하면 tuple로전달
````

````python
def swap(a,b):
    "매개변수순서를바꿔서 return합니다"
    return b,a #여러개값을return하면 tuple로전달
````

````python
result=swap(10,20)
print(result)
print(type(result))
````

````python
x,y=10,20
print(f'함수실행전:x={x},y={y}')
x,y=swap(x,y)
print(f'함수실행후:x={x},y={y}')
````

## 3.3 순서인수,키워드인수
- 순서인수:함수의정의된순서대로전달방식
- 키워드인수:인수의이름을명시하는방식

````python
def func(a,L=None):
    if L is None:
        L = []
    L.append(a)
    return L
````

````python
lst=[]
print(func(10,lst)) #순서인수!
print(func(20)) #순서인수a에20이전달
print(func(20,None))
print(func(L=[],a=10)) #키워드인수
print(func(L=lst,a=10)) #키워드인수
````

## 3.7 인수의언패킹
### 튜플인수언패킹
- 패킹:함수정의시매개변수앞에 *붙임(튜플매개변수)
- 언패킹:함수호출시매개변수앞에*붙임(리스트나튜플이각자요소별매개변수로들어감)

````python
def add(*args):
    tot=0 #누적변수
    for num in args:
        tot+=num
    return tot
````

````python
numbers=(1,2,3,4,5,6)
add(*numbers)
````

````python
num=(1,2,3)
add(*num)
````
