## 5.3 논리연산자: ==> 
and, &, or | ==> 이항연산자(항이두개) /
not ==>일항 연산자

```python
# & : 논리연산자
print(10>3 and 10>5)
print((10>3) and (10>5))
print(10>3 & 10>5)
print((10>3) & (10>5)) # 논리연산 True & True ===> True
print(3 & 10) # 여기서의 &는 논리연산이 아니라 비트연산. 즉 숫자&숫자 는 비트연산함
# 3:0011
# 10:1010
# &:0010 ==> 2
print(10>3 & 10>5) # ==> 그래서 이 경우는 10>2>5 를 의미함. 그래서 false임
```

```python
# and : 논리연산자. 거짓으로 판별되는 첫번째 항의 결과 반환
#       모든항이 참이면 마지막항을 반환
print(True and True and 'abd')
print(0 and 5)
```

```python
# | : 논리연산자
print(10>3 or 10>5)
print((10>3) or (10>5))
print(10>3 | 10>5)
print((10>3) | (10>5)) # 논리연산 True | True ===> True
print(3 | 10) # 여기서의 &는 논리연산이 아니라 비트연산. 즉 숫자&숫자 는 비트연산함
# 3:0011
# 10:1010
# |:1011 ==> 11
print(10>3 | 10>5) # ==> 그래서 이 경우는 10>11>5 를 의미함. 그래서 false임
```

```python
# or : 논리연산자. 참으로 판별되는 첫번째 항의 결과 반환
#       모든항이 거짓이면 마지막항을 반환
print(False or False or '0')
print(0 or 5)
```

- 문자열 슬라이싱변수[from:stop:step]

     - index('찾을문자열'):'찾을문자열'이 왼쪽부터 시작하여 몇번째에 있는지 index를 알려줌(0번째, 1번째...) 없으면 에러가남 ==> 예외처리해야함. 피곤 그래서 find를 써야 함
