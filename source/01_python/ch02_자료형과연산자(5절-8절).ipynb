## 5.1 산술연산자 : +,-,*/ //(몫연산자),%(나머지)

```python
print('10**5(승)=',10**5)
print('4/2(결과 실수)=',4/2)
print('4//2(몫)=', 4//2)
print('4%2(나머지)=',4%2)
```

```python
# 나누기를 제외한 산술연산은 정수끼리 결과는 정수. 정수와 실수 연산결과는 실수
print('10**5.0=',10**5.0)
print('4/2=',4*2)
print('4/2.0=',4*2.0)
```

```python
# 퀴즈1-1. 정수를 입력받아 짝,홀 인지를 여부를 출력 ===> 똑같은 연산 ㅠㅠㅠ
num=int(input('정수를 입력하세요>'))
if num%2:
    temp='홀수'
else:
    temp='짝수'
print('입력하신수는 {}. 2로 나눈 나머지는 {}, 그래서 {}를 입력하셨군요'.format(num,num%2,temp))
print(f'입력하신수는 {num}. 2로 나눈 나머지는 {num%2}, 그래서 {temp}를 입력하셨군요')
```

```python
# 퀴즈1-2. 정수를 입력받아 짝,홀 인지를 여부를 출력 ===> num%2를 result로 미리 정의!!!!! 짧은것보다가독성이 중요!!!
num=int(input('정수를 입력하세요>'))
result=num%2
if num%2:
    temp='홀수'
else:
    temp='짝수'
print('입력하신수는 {}. 2로 나눈 나머지는 {}, 그래서 {}를 입력하셨군요'.format(num,result,temp))
print(f'입력하신수는 {num}. 2로 나눈 나머지는 {result}, 그래서 {temp}를 입력하셨군요')
```

## 5.3 논리연산자: ==> 
and, &, or | ==> 이항연산자(항이두개) /
not ==>일항 연산자

```python
# & : 논리연산자
print(10>3 and 10>5)
print((10>3) and (10>5))
print(10>3 & 10>5)
print((10>3) & (10>5)) # 논리연산 True & True ===> True
print(3 & 10) # 여기서의 &는 논리연산이 아니라 비트연산. 즉 숫자&숫자 는 비트연산함
# 3:0011
# 10:1010
# &:0010 ==> 2
print(10>3 & 10>5) # ==> 그래서 이 경우는 10>2>5 를 의미함. 그래서 false임
```

```python
# and : 논리연산자. 거짓으로 판별되는 첫번째 항의 결과 반환
#       모든항이 참이면 마지막항을 반환
print(True and True and 'abd')
print(0 and 5)
```

```python
# | : 논리연산자
print(10>3 or 10>5)
print((10>3) or (10>5))
print(10>3 | 10>5)
print((10>3) | (10>5)) # 논리연산 True | True ===> True
print(3 | 10) # 여기서의 &는 논리연산이 아니라 비트연산. 즉 숫자&숫자 는 비트연산함
# 3:0011
# 10:1010
# |:1011 ==> 11
print(10>3 | 10>5) # ==> 그래서 이 경우는 10>11>5 를 의미함. 그래서 false임
```

```python
# or : 논리연산자. 참으로 판별되는 첫번째 항의 결과 반환
#       모든항이 거짓이면 마지막항을 반환
print(False or False or '0')
print(0 or 5)
```
