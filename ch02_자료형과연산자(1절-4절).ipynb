```python
def add(x,y=0):
    '''
    매개변수 2개를 받아 두자리 숫자로 만들어요
    x=10의 자리수가 됨
    y=1의자리수가됨(default값0)
    사용법:add(2,3)=>23
    '''
    return x*10+y
```

```python
add(5,4)
```

```python
add(5)
```

```python
a=add
a(5,3)
```
